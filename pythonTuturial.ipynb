### `Python Match`

- The **`match`** statement is used to perform different actions based on different conditions.

- The value **`_`** will always match, so it is important to place it as the last case to make it behave as a default case.


````python
day = 400

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        print("Invalid day")
````

- Use the **`pipe character |`** as an or operator in the case evaluation to check for more than one value match in one case:

````python
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5:
    print("Today is a weekday")
  case 6 | 7:
    print("I love weekends!")
````

- You can add **`if`** statements in the case evaluation as an extra condition-check:

````python
month = 5
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5 if month == 4:
    print("A weekday in April")
  case 1 | 2 | 3 | 4 | 5 if month == 5:
    print("A weekday in May")
  case _:
    print("No match")
````

## `yield` vs `return` Summary

### `return`
- Returns **all data at once**.
- Creates the entire collection in memory before returning it.
- Best for **small datasets**.

```python
def count():
    return [1, 2, 3, 4, 5]
```

---

### `yield`
- Returns **one value at a time**.
- Does **not** store all values in memory.
- Produces values **only when needed** (lazy evaluation).
- Best for **large datasets**, files, database records, and data streams.

```python
def count():
    for i in range(1, 6):
        yield i
```

---

## Advantages of `yield`

- ✅ Uses much less memory.
- ✅ Generates values on demand.
- ✅ Can handle very large or even infinite sequences.
- ✅ Stops and resumes execution automatically.
- ✅ Improves performance when you don't need all values.

---

## Comparison

| `return` | `yield` |
|----------|---------|
| Returns all values at once | Returns one value at a time |
| Higher memory usage | Lower memory usage |
| Function ends after `return` | Function pauses and resumes |
| Good for small data | Good for large or streaming data |

### Conclusion
Use **`return`** when you need all results immediately.  
Use **`yield`** when working with large amounts of data or when values should be generated only as they are needed.

# Generator Methods

### `next()`
- Starts or resumes the generator.
- Returns the next value from `yield`.
- Pauses again at the next `yield`.

---

### `send(value)`
- Sends a value into the generator.
- The value is received by the current `yield` expression.
- The generator must be started first with `next()` (or `send(None)`).

---

### `close()`
- Stops the generator immediately.
- Executes any `finally` block before closing.
- Further calls to `next()` raise `StopIteration`.

---

## Summary

| Method | Purpose |
|---------|---------|
| `next()` | Get the next value. |
| `send(value)` | Send a value into the generator. |
| `close()` | Stop the generator. |

## 2. The Main Classes in `datetime`

The `datetime` module provides several classes for working with dates and times.

| Class | Purpose |
|--------|---------|
| `date` | Represents a date (`year`, `month`, `day`). |
| `time` | Represents a time (`hour`, `minute`, `second`, `microsecond`). |
| `datetime` | Combines both **date** and **time** into a single object. |
| `timedelta` | Represents a duration or the difference between two dates/times. |
| `timezone` | Represents time zone information. |

> **Note:** In this course, we'll focus on the most commonly used classes: `date`, `time`, `datetime`, and `timedelta`.

- *Getting the Current Date and Time*

````python
print(datetime.datetime.now())
````

- *Getting Individual Components*

````python
now = datetime.datetime.now()

print(now.year) #! print Year only
print(now.month) #! print Month only
print(now.day) #! print day only
print(now.hour) #! print hours only
print(now.minute) #! print minutes only 
print(now.second) #! print seconds only
print(now.microsecond) #! print microseconds only
````

- **Formatting Dates (strftime)**
    - *strftime() converts a date or datetime object into a formatted string.*

#### Common `strftime` / `strptime` Format Codes

These format codes are used to format or parse dates and times.

| Code | Meaning | Example |
|------|---------|---------|
| `%Y` | 4-digit year | `2026` |
| `%y` | 2-digit year | `26` |
| `%m` | Month (`01`–`12`) | `07` |
| `%d` | Day of the month | `12` |
| `%H` | Hour (24-hour format) | `13` |
| `%I` | Hour (12-hour format) | `01` |
| `%M` | Minute | `45` |
| `%S` | Second | `20` |
| `%A` | Full weekday name | `Sunday` |
| `%a` | Abbreviated weekday name | `Sun` |
| `%B` | Full month name | `July` |
| `%b` | Abbreviated month name | `Jul` |
| `%p` | AM/PM | `PM` |

````python
from datetime import datetime

now = datetime.now()

print(now.strftime("%Y-%m-%d"))

print(now.strftime("%y-%m-%d"))

print(now.strftime("%Y-%m-%d %H-%M-%S")) #! H : 24 H
print(now.strftime("%Y-%m-%d %I-%M-%S")) #! I : 12 H

print(now.strftime("%Y-%m-%d %A %H-%M-%S"))
print(now.strftime("%Y-%m-%d %a %H-%M-%S"))
print(now.strftime("%Y-%m-%d %B/%A %H-%M-%S"))
print(now.strftime("%Y-%m-%d %b/%A %H-%M-%S"))
print(now.strftime("%Y-%m-%d %b/%A %H-%M-%S %p"))
````

- **Converting Strings to Dates (strptime)**
    - *strptime() parses a string into a datetime object*

````python
from datetime import datetime

text = "12/07/2026"
d1 = datetime.strptime(text, "%d/%m/%Y")
print(d1)

text2 = "2026-05-13"
d2 = datetime.strptime(text2, "%Y-%m-%d")
print(d2)

text3 = "26-05-13"
d3 = datetime.strptime(text3, "%y-%m-%d")
print(d3)

text4 = "2026-05-13 15:10:30"
d4 = datetime.strptime(text4, "%Y-%m-%d %H:%M:%S")
print(d4)
````

````python
from datetime import datetime

text = "26-05-13"

d = datetime.strptime(text, "%y-%m-%d")

print(d.strftime("%y-%m-%d"))
````
