### `Python Match`

- The **`match`** statement is used to perform different actions based on different conditions.

- The value **`_`** will always match, so it is important to place it as the last case to make it behave as a default case.


```python
day = 400

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        print("Invalid day")
```

- Use the **`pipe character |`** as an or operator in the case evaluation to check for more than one value match in one case:

```python
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5:
    print("Today is a weekday")
  case 6 | 7:
    print("I love weekends!")
```

- You can add **`if`** statements in the case evaluation as an extra condition-check:

```python
month = 5
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5 if month == 4:
    print("A weekday in April")
  case 1 | 2 | 3 | 4 | 5 if month == 5:
    print("A weekday in May")
  case _:
    print("No match")
```

### `Recursion`

```python
def countDown(n):
    if n == 0: # base case
        return
    print(n)
    countDown(n - 1) # recursive case

countDown(5)
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

fact = factorial(5)
print(fact)
```

```python
def summation(n):
    if n == 0:
        return 0
    return n + summation(n - 1)

sum = summation(5)
print(sum)
```

```python
word = "WWWrrreeeesss"
def cleanString(word):
    if len(word) == 1:
        return word
    if word[0] == word[1]:
        return cleanString(word[1:])
    return word[0] + cleanString(word[1:])

result = cleanString(word)
print(result)
```

```python
word = "hello"

def reverseWord(word):
    if len(word) == 1:
        return word
    return word[-1] + reverseWord(word[:-1])

result = reverseWord(word)
print(result)
```

- **Filter(iterable, function)**

```python
def checkNum(n):
    if n > 10:
        return n
    
nums = [1,19,5,10,3,15,100]

result = filter(checkNum, nums)

print(result)
for i in result:
    print(i)
```

```python
def checkZero_v1(n):
    if n == 0:
        return n
    
nums = [0,0,0,1,5,9]

r = filter(checkZero_v1, nums)
for i in r:
    print(i)

#! No result found as 0 is false and filter required that the result should be true
```

```python
def checkZero_v2(n):
    if n == 0:
        return True

nums = [0,0,0,1,5,9]

r = filter(checkZero_v2, nums)
for i in r:
    print(i)
```
